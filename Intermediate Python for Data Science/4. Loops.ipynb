Ada beberapa teknik yang dapat Anda gunakan untuk berulang kali mengeksekusi kode Python. **While loops** seperti pernyataan `if` yang diulangi, `for loop iterates` atas semua jenis struktur data. Pelajari semua tentang mereka di bab ini.

## while loop

### while: warming up

While loop seperti pernyataan `if` yang diulangi. Kode dieksekusi berulang-ulang, selama kondisinya Benar.

<pre>
while condition :
    expression
</pre>



```python
x = 1
while x < 4 :
    print(x)
    x = x + 1
```

**Note** : Setelah 3 berjalan, `x` akan sama dengan 4, menyebabkan `x < 4` untuk mengevaluasi ke `False`. Ini berarti bahwa while loop dieksekusi 3 kali, memberikan tiga hasil cetak.

### Basic while loop

```python
# Initialize offset
offset = 8

# Code the while loop
while offset != 0 :
    print("correcting...")
    offset = offset - 1
    print(offset)
```

### Add conditionals

`While` loop yang mengoreksi `offset` adalah awal yang baik, tetapi bagaimana jika `offset` negatif? Anda dapat mencoba menjalankan kode berikut ini di mana `offset` diinisialisasi ke `-6` :

<pre>
# Initialize offset
offset = -6

# Code the while loop
while offset != 0 :
    print("correcting...")
    offset = offset - 1
    print(offset)
</pre>

Tetapi sesi Anda akan terputus. `while` loop tidak akan pernah berhenti berjalan, karena `offset` akan semakin menurun pada setiap putaran. `offset != 0` tidak akan pernah menjadi `False` dan `while` loop berlanjut selamanya.

Perbaiki beberapa hal dengan menambahkan pernyataan `if-else` di dalam `while` loop. Jika kode Anda masih terlalu lama untuk dijalankan, Anda mungkin membuat kesalahan!

```python
# Initialize offset
offset = -6

# Code the while loop
while offset != 0 :
    print("correcting...")
    if offset > 0 :
      offset -= 1
    else : 
      offset += 1    
    print(offset)
```

**Note** : `while` loop tidak sering digunakan dalam Ilmu Data, jadi mari kita beralih ke `for` loop.

## for loop

### Loop over a list

Berikut ini adalah contoh kode `for` loop:

```python
fam = [1.73, 1.68, 1.71, 1.89]
for height in fam : 
    print(height)
```

```python
# areas list
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Code the for loop
for area in areas :
    print(area)
```

### Indexes and values

Menggunakan `for` loop untuk iterasi list hanya memberi Anda akses ke setiap elemen list dalam setiap proses, satu demi satu. Jika Anda juga ingin mengakses informasi indeks, jadi di mana elemen list yang Anda iterasi berada, Anda dapat menggunakan `enumerate()`.

```python
fam = [1.73, 1.68, 1.71, 1.89]
for index, height in enumerate(fam) :
    print("person " + str(index) + ": " + str(height))
```

```python
# areas list
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Change for loop to use enumerate() and update print()
for x, y in enumerate(areas) :
    print("room " + str(x) + ": " + str(y))
```

Untuk orang yang bukan programmer, `room 0: 11.25` tampak aneh. Bukankah lebih baik jika hitungan dimulai dari 1?

```python
# areas list
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Code the for loop
for index, area in enumerate(areas) :
    print("room " + str(index+1) + ": " + str(area))
```

### Loop over list of lists

Ingat variabel `house` dari kursus Intro Python? Itu pada dasarnya adalah `lists of lists`, di mana setiap sublist berisi nama dan area kamar di rumah Anda.

```python
# house list of lists
house = [["hallway", 11.25], 
         ["kitchen", 18.0], 
         ["living room", 20.0], 
         ["bedroom", 10.75], 
         ["bathroom", 9.50]]
         
# Build a for loop from scratch
for x, y in house :
    print("the " + str(x) + " is " + str(y) + " sqm")
```

## Looping Data Structures

### Loop over dictionary

Di Python 3, Anda memerlukan metode `items()` untuk melakukan loop pada dictionary :

```python
world = { "afghanistan":30.55, 
          "albania":2.77,
          "algeria":39.21 }

for key, value in world.items() :
    print(key + " -- " + str(value))
```

Masih ingatkah dictionary `eropa` yang memuat nama beberapa negara Eropa sebagai `key` dan ibukotanya sebagai `value` yang sesuai? Teruskan dan tulis satu lingkaran untuk mengulanginya!

```python
# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin',
          'norway':'oslo', 'italy':'rome', 'poland':'warsaw', 'austria':'vienna' }
          
# Iterate over europe
for k, v in europe.items() :
    print("the capital of " + str(k) + " is " + str(v))
```

**Note** : Perhatikan bahwa urutan hasil cetak tidak selalu sesuai dengan urutan yang digunakan saat mendefinisikan `europe`. Ingat: dictionary pada dasarnya tidak berurutan!

### Loop over DataFrame

Iterating melalui Pandas DataFrame dapat dilakukan dengan metode `iterrows()`. Digunakan dalam `for` loop, setiap pengamatan diulang berulang dan pada setiap iterasi label baris dan isi baris aktual tersedia:

<pre>
for lab, row in brics.iterrows() :
    ...
</pre>

```python
# Import cars data
import pandas as pd
cars_data = 'https://assets.datacamp.com/production/repositories/287/datasets/79b3c22c47a2f45a800c62cae39035ff2ea4e609/cars.csv'
cars = pd.read_csv(cars_data, index_col = 0)

# Iterate over rows of cars
for lab, row in cars.iterrows() :
    print(lab)
    print(row)
```

Data baris yang dihasilkan oleh `iterrows()` di setiap proses adalah Pandas Series. Format ini sangat tidak nyaman untuk dicetak. Untungnya, Anda dapat dengan mudah memilih variabel dari Pandas Series menggunakan tanda kurung siku:

<pre>
for lab, row in brics.iterrows() :
    print(row['country'])
</pre>

```python
# Adapt for loop
for lab, row in cars.iterrows() :
    print(lab + ": " + str(row["cars_per_cap"]))
```

### Add column

Contoh cara menambahkan panjang nama negara dari DataFrame brics di kolom baru:

<pre>
for lab, row in brics.iterrows() :
    brics.loc[lab, "name_length"] = len(row["country"])
</pre>

Anda dapat melakukan hal serupa pada DataFrame `cars`.

```python
# Code for loop that adds COUNTRY column
for lab, row in cars.iterrows() :
    cars.loc[lab, "COUNTRY"] = row["country"].upper()

# Print cars
print(cars)
```

Menggunakan `iterrows()` untuk melakukan iterasi pada setiap pengamatan Pandas DataFrame mudah dipahami, tetapi tidak terlalu efisien. Di setiap iterasi, Anda membuat Pandas Series baru.

Jika Anda ingin menambahkan kolom ke DataFrame dengan memanggil fungsi pada kolom lain, metode `iterrows()` dalam kombinasi dengan `for` loop bukan cara yang lebih disukai. Sebagai gantinya, Anda dapat menggunakan `apply()`.

Bandingkan versi `iterrows()` dengan versi `apply()` untuk mendapatkan hasil yang sama dalam DataFrame `brics` :

<pre>
for lab, row in brics.iterrows() :
    brics.loc[lab, "name_length"] = len(row["country"])

brics["name_length"] = brics["country"].apply(len)
</pre>

Kita dapat melakukan hal serupa untuk memanggil metode `upper()` pada setiap nama di kolom `country`. Namun, `upper()` adalah metode, jadi kita membutuhkan pendekatan yang sedikit berbeda:

```python
# Use .apply(str.upper)
# for lab, row in cars.iterrows() :
    # cars.loc[lab, "COUNTRY"] = row["country"].upper()
    
cars["COUNTRY"] = cars["country"].apply(str.upper)
print(cars)
```
