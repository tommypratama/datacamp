## Histograms

Untuk melihat bagaimana harapan hidup di berbagai negara didistribusikan, mari kita buat histogram `life_exp`.

```python
# Create histogram of life_exp data
plt.hist(life_exp)

# Display histogram
plt.show()
```

Pada histogram di atas, Anda tidak menentukan jumlah `bins`. Secara default, Python menetapkan jumlah `bins` ke 10. Jumlah `bins` cukup penting. Terlalu sedikit `bins` akan terlalu menyederhanakan kenyataan dan tidak akan menunjukkan detailnya kepada Anda. Terlalu banyak `bins` akan memperumit kenyataan dan tidak akan menampilkan gambar yang lebih besar.

Untuk mengontrol jumlah `bins` untuk membagi data Anda, Anda dapat mengatur argumen `bins`.

Itulah tepatnya yang akan Anda lakukan dalam latihan ini. Anda akan membuat dua plot di sini. Kode dalam skrip sudah termasuk panggilan `plt.show()` dan `plt.clf()`; `plt.show()` menampilkan plot; `plt.clf()` membersihkannya lagi sehingga Anda dapat memulai dari awal.

```python
# Build histogram with 5 bins
plt.hist(life_exp, bins=5)

# Show and clean up plot
plt.show()
plt.clf()

# Build histogram with 20 bins
plt.hist(life_exp, bins=20)

# Show and clean up again
plt.show()
plt.clf()
```

## Customization

### Labels

Saatnya menyesuaikan plot Anda sendiri. Ini adalah bagian yang menyenangkan, Anda akan melihat plot Anda menjadi hidup!

Sebagai langkah pertama, mari tambahkan label sumbu dan judul ke plot. Anda dapat melakukan ini dengan fungsi `xlabel()`, `ylabel()` dan `title()` tersedia di matplotlib.pyplot. Sub-paket ini sudah diimpor sebagai `plt`.

```python
# Basic scatter plot, log scale
plt.scatter(gdp_cap, life_exp)
plt.xscale('log') 

# Strings
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)

# Add title
plt.title(title)

# After customizing, display the plot
plt.show()
```

### Ticks

Dalam contoh ini, `ticks` yang sesuai dengan angka 0, 1 dan 2 akan diganti masing-masing dengan `one`, `two` dan `three`.

<pre>plt.yticks([0,1,2], ["one","two","three"])</pre>

Mari kita lakukan hal serupa untuk sumbu x dari *world development chart* Anda, dengan fungsi `xticks()` . Nilai **ticks** yaitu `1000`, `10000` dan `100000` harus diganti dengan `1k`, `10k` dan `100k`. Untuk tujuan ini, buat dua list yang terdiri dari: `tick_val` dan `tick_lab`.

```python
# Scatter plot
plt.scatter(gdp_cap, life_exp)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

# Definition of tick_val and tick_lab
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val, tick_lab)

# After customizing, display the plot
plt.show()
```

### Sizes

Saat ini, scatter plot hanyalah awan titik-titik biru, tidak bisa dibedakan satu sama lain. Mari kita ubah ini. Bukankah lebih baik jika ukuran titik-titik sesuai dengan populasi?

Anda dapat menambahkan list `pop` ke metode scatter, sebagai argumen `s`, yaitu untuk *size*.

```python
# Import numpy as np
import numpy as np

# Store pop as a numpy array: np_pop
np_pop = np.array(pop)

# Double np_pop
np_pop = np_pop * 2

# Update: set s argument to np_pop
plt.scatter(gdp_cap, life_exp, s = np_pop)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])

# Display the plot
plt.show()
```

Kebanyakan orang tinggal di negara-negara berpenghasilan menengah di mana perbedaan umur sangat besar di antara negara-negara; tergantung pada bagaimana pendapatan didistribusikan dan bagaimana digunakan.

Bagaimana anda menggambarkan tentang plot di atas?

* **Jawab** : Negara-negara dengan warna biru, sesuai dengan Afrika, memiliki harapan hidup yang rendah dan PDB per kapita yang rendah.
