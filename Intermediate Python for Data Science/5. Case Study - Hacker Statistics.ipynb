### Determine your next move

Dalam taruhan *Empire State Building*, langkah Anda berikutnya tergantung pada jumlah mata yang Anda lemparkan dengan dadu. Kita dapat mengkodekan ini dengan sempurna menggunakan konstruksi `if` - `elif` - `else`!

```python
# Numpy is imported, seed is set

# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1, 7)

# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice <= 5 :
    step = step - 1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice)
print(step)
```

**Note** : Anda melemparkan 6, sehingga kode untuk pernyataan lain dieksekusi. Anda melempar lagi, dan ternyata Anda melempar 3, menyebabkan Anda naik tiga langkah: Anda saat ini pada langkah 53.

## Random Walk

### The next step

Sebelumnya, Anda sudah menulis kode Python yang menentukan langkah selanjutnya berdasarkan langkah sebelumnya. Sekarang saatnya untuk meletakkan kode ini di dalam `for` loop sehingga kita dapat mensimulasikan `random walk`.

```python
# Numpy is imported, seed is set

# Initialize random_walk
random_walk = [0]

# Complete the ___
for x in range(100) :
    # Set step: last element in random_walk
    step = random_walk[-1]

    # Roll the dice
    dice = np.random.randint(1,7)

    # Determine next step
    if dice <= 2:
        step = step - 1
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    # append next_step to random_walk
    random_walk.append(step)

# Print random_walk
print(random_walk)
```

**Note** : Masih ada yang salah: terdapat indeks negatif!

### How low can you go?

Segalanya berubah dengan baik! Anda sudah memiliki kode yang menghitung lokasi Anda di Empire State Building setelah 100 lemparan dadu. Namun, ada sesuatu yang belum kami pikirkan - Anda tidak bisa pergi di bawah 0!

Cara khas untuk menyelesaikan masalah seperti ini adalah dengan menggunakan `max()`. Jika Anda melewati `max()` dua argumen, argumen terbesar akan dikembalikan. Misalnya, untuk memastikan bahwa variabel `x` tidak pernah di bawah `10` ketika Anda menguranginya, Anda dapat menggunakan:

<pre>x = max(10, x - 1)</pre>

```python
# Numpy is imported, seed is set

# Initialize random_walk
random_walk = [0]

for x in range(100) :
    step = random_walk[-1]
    dice = np.random.randint(1,7)

    if dice <= 2:
        # Replace below: use max to make sure step can't go below 0
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)

    random_walk.append(step)

print(random_walk)
```

**Note** : Jika Anda melihat dekat pada output, Anda akan melihat bahwa di sekitar indeks 15 langkahnya tetap pada 0. Anda tidak akan di bawah nol lagi. (Tidak sama dengan output datacamp! karena angka selalu acak)

## Distribution

Satu `random walk` adalah satu hal, tetapi itu tidak memberi tahu Anda jika Anda memiliki peluang bagus untuk memenangkan taruhan.

Untuk mendapatkan ide tentang seberapa besar peluang Anda mencapai 60 langkah, Anda dapat berulang kali mensimulasikan `random walk` dan mengumpulkan hasilnya. Itulah tepatnya yang akan Anda lakukan dalam latihan ini.

```python
# Numpy is imported; seed is set

# Initialize all_walks (don't change this line)
all_walks = []

# Simulate random walk 10 times
for i in range(10) :

    # Code from before
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)

        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)

    # Append random_walk to all_walks
    all_walks.append(random_walk)

# Print all_walks
print(all_walks)
```

### Visualize all walks

`all_walks` adalah lists of lists : setiap sub-list mewakili satu random walk. Jika Anda mengonversi lists of lists ini menjadi array Numpy, Anda dapat mulai membuat plot menarik! 

```python
# numpy and matplotlib imported, seed set.

# initialize and populate all_walks
all_walks = []
for i in range(10) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to Numpy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()

# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
# Sekarang setiap baris di np_all_walks mewakili posisi setelah 1 lemparan untuk 10 random walk.
np_aw_t = np.transpose(np_aw)

# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

**Note** : Anda dapat dengan jelas melihat bagaimana berbagai simulasi `random walk` berjalan. Transposing array 2D Numpy sangat penting; jika tidak Python dapat disalahpahami.

### Implement clumsiness

Dengan kode Anda yang ditulis dengan rapi ini, mengubah berapa kali `random walk` untuk disimulasikan sangatlah mudah. Anda cukup memperbarui fungsi `range()` di level atas pada `for` loop.

Masih ada sesuatu yang kita lupa! Anda sedikit canggung dan Anda memiliki peluang 0,1% untuk jatuh. Itu panggilan untuk generate nomor acak lain. Pada dasarnya, Anda dapat menghasilkan float acak antara `0` dan `1`. Jika nilai ini kurang dari atau sama dengan 0,001, Anda harus mengatur ulang langkah ke 0.

```python
# numpy and matplotlib imported, seed set

# Simulate random walk 250 times
all_walks = []
for i in range(250) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # Implement clumsiness
        if np.random.rand() <= 0.001 :
            step = 0

        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))
plt.plot(np_aw_t)
plt.show()
```

**Note** : Lihatlah plotnya. Dalam beberapa dari 250 simulasi Anda memang benar-benar menyelam!

### Plot the distribution

Semua visualisasi mewah ini telah menempatkan kita pada jalur yang salah. Kami masih harus menyelesaikan masalah jutaan dolar: **Berapa peluang Anda akan mencapai 60 langkah di Empire State Building?**

Pada dasarnya, Anda ingin tahu tentang titik akhir dari semua `random walk` yang telah disimulasikan. Titik akhir ini memiliki distribusi tertentu yang dapat Anda visualisasikan dengan histogram.

```python
# numpy and matplotlib imported, seed set

# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np_aw_t[-1]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```

**Note** : Lihatlah histogram; Menurut Anda, apa peluang Anda?

### Calculate the odds

Histogram dari latihan sebelumnya dibuat dari array Numpy `ends`, yang berisi 500 bilangan bulat. Setiap bilangan bulat mewakili titik akhir dari `random walk`. Untuk menghitung kemungkinan bahwa titik akhir ini lebih besar dari atau sama dengan 60, Anda dapat menghitung jumlah bilangan bulat `ends` yang lebih besar dari atau sama dengan 60 dan membagi angka itu dengan 500, jumlah total simulasi.

Kalau begitu, berapa perkiraan perkiraan Anda akan mencapai 60 langkah jika Anda memainkan game Empire State Building ini? Array `ends` adalah semua yang Anda butuhkan;

```python
np.mean(ends >= 60)
```
