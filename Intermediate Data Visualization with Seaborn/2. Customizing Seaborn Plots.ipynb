## Using Seaborn Styles

### Setting the default style

Untuk latihan-latihan ini, kita akan melihat nilai sewa pasar pameran yang dihitung oleh Departemen Pembangunan Perumahan dan Perkotaan AS. Data ini digunakan untuk menghitung pedoman untuk beberapa program federal. Nilai aktual untuk sewa sangat bervariasi di seluruh AS. Kami dapat menggunakan data ini untuk mendapatkan pengalaman dengan mengonfigurasi plot Seaborn.

```python
# import all modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Read in the DataFrame
df = pd.read_csv("datasets/FY18_4050_FMRs.csv")

# Print data
df.head()
```

```python
# Plot the pandas histogram
df['fmr_2'].plot.hist()
plt.show()
plt.clf()

# Set the default seaborn style
sns.set()

# Plot the pandas histogram again
df['fmr_2'].plot.hist()
plt.show()
plt.clf()
```

**Note** : Anda dapat melihat bagaimana histogram diubah ketika gaya Seaborn diterapkan pada visualisasi yang mendasarinya.

### Comparing styles

Seaborn mendukung pengaturan gaya yang berbeda yang dapat mengontrol estetika plot akhir. Dalam latihan ini, Anda akan memplot data yang sama dalam dua gaya yang berbeda untuk melihat bagaimana gaya mengubah output.

```python
sns.set_style("dark")

sns.distplot(df["fmr_2"])

plt.show()

plt.clf()
```

```python
sns.set_style("whitegrid")

sns.distplot(df["fmr_2"])

plt.show()

plt.clf()
```

**Note** : Gaya Seaborn menyediakan cara cepat dan mudah untuk mengubah visualisasi Anda secara konsisten dan menarik secara visual.

### Removing spines

Secara umum, visualisasi harus meminimalkan tanda asing sehingga data berbicara dengan sendirinya. Seaborn memungkinkan Anda untuk menghapus garis-garis di sumbu atas, bawah, kiri dan kanan, yang sering disebut *spines*.

```python
# Set the style to white
sns.set_style('white')

# Create a regression plot
sns.lmplot(data=df,
           x='pop2010',
           y='fmr_2')

# Remove the spines
sns.despine()

# Show the plot and clear the figure
plt.show()
plt.clf()
```

**Note** : Menghapus *spines* memungkinkan data berbicara sendiri.

## Colors in Seaborn

### Matplotlib color codes

Seaborn menawarkan beberapa opsi untuk memodifikasi warna visualisasi Anda. Pendekatan paling sederhana adalah dengan secara eksplisit menyatakan warna plot. Cara cepat untuk mengubah warna adalah dengan menggunakan kode warna standar `matplotlib`.

```python
# Set style, enable color code, and create a magenta distplot
sns.set(color_codes=True)
sns.distplot(df['fmr_3'], color='m')

# Show the plot
plt.show()
```

**Note** : Ini adalah cara cepat untuk mengubah warna plot.

### Using default palettes

Seaborn menyertakan beberapa palet default yang dapat dengan mudah diterapkan pada plot Anda. Dalam contoh ini, kita akan melihat dampak dari dua palet yang berbeda pada `distplot` yang sama.

```python
# Loop through differences between bright and colorblind palettes
for p in ['bright', 'colorblind']:
    sns.set_palette(p)
    sns.distplot(df['fmr_3'])
    plt.show()
    
    # Clear the plots    
    plt.clf()
```

**Note** : Menggunakan palet buta warna (*colorblind*) default adalah pilihan yang baik untuk memastikan visualisasi Anda mudah dibaca untuk individu dengan derajat buta warna.

### Creating Custom Palettes

Memilih palet yang kohesif yang berfungsi untuk data Anda bisa menghabiskan waktu. Untungnya, Seaborn memiliki beberapa fungsi yang memungkinkan Anda untuk membuat palet sekuensial, kategoris, atau divergen kustom Anda sendiri. Seaborn juga memudahkan untuk melihat palet Anda dengan menggunakan fungsi `palplot()`.

Dalam latihan ini, Anda dapat bereksperimen dengan membuat palet berbeda.

```python
# Create a Purples sequential palette containing 8 colors.
sns.palplot(sns.color_palette("Purples", 8))

# Display plot
plt.show()
```

```python
# Create a palette with 10 colors using the husl system.
sns.palplot(sns.color_palette("husl", 10))

# Display plot
plt.show()
```

```python
# Create a diverging palette with 6 colors coolwarm.
sns.palplot(sns.color_palette("coolwarm", 6))

# Display plot
plt.show()
```

**Note** : Kemungkinan palet warna tidak terbatas. Menggunakan bantuan fungsi-fungsi ini akan memungkinkan Anda untuk membuat plot unik dan menarik secara visual yang akan menonjol dan menyampaikan poin Anda.

## Customizing with matplotlib

### Using matplotlib axes

Seaborn menggunakan matplotlib sebagai pustaka dasar untuk membuat plot. Sebagian besar waktu, Anda dapat menggunakan Seaborn API untuk memodifikasi visualisasi Anda, tetapi kadang-kadang menggunakan fungsi matplotlib membantu untuk menyesuaikan plot Anda. Objek paling penting dalam kasus ini adalah sumbu (`axes`) matplotlib.

Setelah Anda memiliki objek `axes`, Anda dapat melakukan banyak penyesuaian plot Anda.

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of data
sns.distplot(df['fmr_3'], ax=ax)

# Create a more descriptive x axis label
ax.set(xlabel="3 Bedroom Fair Market Rent")

# Show the plot
plt.show()
```

**Note** : Sekarang Anda tahu dasar-dasar cara menyesuaikan plot.

### Additional plot customizations

API `matplotlib` mendukung banyak penyesuaian umum seperti pelabelan sumbu, penambahan judul, dan batas pengaturan. Mari kita selesaikan latihan penyesuaian lainnya.

```python
# Create a figure and axes
fig, ax = plt.subplots()

# Plot the distribution of 1 bedroom rents
sns.distplot(df['fmr_1'], ax=ax)

# Modify the properties of the plot
ax.set(xlabel="1 Bedroom Fair Market Rent",
       xlim=(100, 1500),
       title="US Rent")

# Display the plot
plt.show()
```

**Note** : Membuat jenis penyesuaian ini untuk plot Anda benar-benar dapat meningkatkan dampaknya.

### Adding annotations

Setiap perangkat tambahan yang telah kita bahas dapat digabungkan bersama. Dalam latihan berikutnya, kita dapat membuat anotasi plot distribusi untuk menyertakan garis yang menunjukkan harga sewa rata-rata dan median.

Untuk contoh ini, palet telah diubah menjadi `bright` menggunakan `sns.set_palette()`

```python
# Set palette
sns.set_palette("bright")
```

```python
# Create a figure and axes. Then plot the data
fig, ax = plt.subplots()
sns.distplot(df['fmr_1'], ax=ax)

# Customize the labels and limits
ax.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500), title="US Rent")

# Add vertical lines for the median and mean
ax.axvline(x=df['fmr_1'].median(), color='m', label='Median', linestyle='--', linewidth=2)
ax.axvline(x=df['fmr_1'].mean(), color='b', label='Mean', linestyle='-', linewidth=2)

# Show the legend and plot the data
ax.legend()
plt.show()
```

**Note** : Itu adalah contoh yang panjang tetapi harus menunjukkan kekuatan yang tersedia untuk Anda!

### Multiple plots

Untuk latihan terakhir, kita akan memplotting perbandingan sewa pasar pameran untuk apartemen 1 kamar tidur dan 2 kamar tidur.

```python
# Create a plot with 1 row and 2 columns that share the y axis label
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, sharey=True)

# Plot the distribution of 1 bedroom apartments on ax0
sns.distplot(df['fmr_1'], ax=ax0)
ax0.set(xlabel="1 Bedroom Fair Market Rent", xlim=(100,1500))

# Plot the distribution of 2 bedroom apartments on ax1
sns.distplot(df['fmr_2'], ax=ax1)
ax1.set(xlabel="2 Bedroom Fair Market Rent", xlim=(100,1500))

# Display the plot
plt.show()
```
