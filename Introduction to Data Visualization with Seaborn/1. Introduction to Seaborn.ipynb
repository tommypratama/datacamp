### Making a count plot with a DataFrame

Dalam latihan ini, kita akan melihat respons terhadap survei yang dikirimkan kepada kaum muda. Pertanyaan utama kami di sini adalah: berapa banyak anak muda yang disurvei melaporkan ketakutan terhadap laba-laba? ## Adding a third variable with hue

### Hue and scatter plots

Dalam video sebelumnya, kami belajar bagaimana rona (`hue`) memungkinkan kami untuk dengan mudah membuat subkelompok dalam plot Seaborn. Mari kita coba dengan mengeksplorasi data dari siswa di sekolah menengah. Kami memiliki banyak informasi tentang setiap siswa seperti usia mereka, di mana mereka tinggal, kebiasaan belajar mereka dan kegiatan ekstrakurikuler mereka.

Untuk saat ini, kita akan melihat hubungan antara jumlah absen yang mereka miliki di sekolah dan nilai akhir mereka dalam kursus, tersegmentasi berdasarkan tempat tinggal siswa (daerah pedesaan vs perkotaan).

```python
# Load data
student_data = pd.read_csv('datasets/student-alcohol-consumption.csv', index_col=0)
```

```python
# Create a scatter plot of absences vs. final grade
sns.scatterplot(x='absences', y='G3', data=student_data, hue='location')

# Show plot
plt.show()
```

Buat `"Rural"` muncul sebelum `"Urban"` dalam legenda pada plot.

```python
# Change the legend order in the scatter plot
sns.scatterplot(x="absences", y="G3", 
                data=student_data, 
                hue="location",
                hue_order=["Rural", "Urban"])

# Show plot
plt.show()
```

**Note** : Sepertinya siswa dengan absensi yang lebih tinggi cenderung memiliki nilai lebih rendah di daerah pedesaan dan perkotaan.

### Hue and count plots

Mari kita lanjutkan mengeksplorasi dataset siswa di sekolah menengah dengan melihat variabel baru. Kolom `"school"` menunjukkan inisial sekolah yang dihadiri siswa - "GP" atau "MS".

Dalam latihan terakhir, kita membuat scatter plot di mana titik plot diwarnai berdasarkan apakah siswa tinggal di daerah perkotaan atau pedesaan. Berapa banyak siswa yang tinggal di daerah perkotaan vs pedesaan, dan apakah ini bervariasi berdasarkan sekolah apa yang dihadiri siswa? Mari kita buat plot perhitungan dengan subkelompok untuk mengetahuinya.

```python
# Create a dictionary mapping subgroup values to colors
palette_colors = {"Rural": "green", "Urban": "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x="school", data=student_data, hue="location", palette=palette_colors)

# Display plot
plt.show()
```

**Note** : Siswa di GP cenderung berasal dari lokasi perkotaan, tetapi siswa di MS lebih merata.
