```python
# Import library
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# Load data
survey_data = pd.read_csv("datasets/young-people-survey-4.csv")
survey_data.head()
```

### Changing style and palette

Mari kembali ke dataset kami yang berisi hasil survei yang diberikan kepada kaum muda tentang kebiasaan dan preferensi mereka. Kami telah menyediakan kode untuk membuat count plot tanggapan mereka terhadap pertanyaan *"Seberapa sering Anda mendengarkan saran orang tua Anda?"*. Sekarang mari kita ubah gaya dan palet untuk membuat alur ini lebih mudah untuk ditafsirkan.

```python
# Set the style to "whitegrid"
sns.set_style("whitegrid")

# Create a count plot of survey responses
category_order = ["Never", "Rarely", "Sometimes", 
                  "Often", "Always"]

sns.catplot(x="Parents Advice", 
            data=survey_data, 
            kind="count", 
            order=category_order)

# Show plot
plt.show()
```

```python
# Set the color palette to "Purples"
sns.set_style("whitegrid")
sns.set_palette("Purples")

# Create a count plot of survey responses
category_order = ["Never", "Rarely", "Sometimes", 
                  "Often", "Always"]

sns.catplot(x="Parents Advice", 
            data=survey_data, 
            kind="count", 
            order=category_order)

# Show plot
plt.show()
```

```python
# Change the color palette to "RdBu"
sns.set_style("whitegrid")
sns.set_palette("RdBu")

# Create a count plot of survey responses
category_order = ["Never", "Rarely", "Sometimes", 
                  "Often", "Always"]

sns.catplot(x="Parents Advice", 
            data=survey_data, 
            kind="count", 
            order=category_order)

# Show plot
plt.show()
```

**Note** : Gaya dan palet warna yang berbeda ini menyoroti perbedaan antara jumlah anak muda yang biasanya mendengarkan saran orang tua mereka dibandingkan dengan mereka yang tidak.

### Changing the scale

Apakah persentase orang yang melaporkan bahwa mereka merasa kesepian berbeda-beda tergantung pada berapa banyak saudara kandung yang mereka miliki? Mari cari tahu menggunakan plot bar, sambil juga mengeksplorasi empat skala plot Seaborn yang berbeda ("contexts").

```python
# Set the context to "paper"
sns.set_context("paper")

# Create bar plot
sns.catplot(x="Number of Siblings", y="Feels Lonely",
            data=survey_data, kind="bar")

# Show plot
plt.show()
```

```python
# Change the context to "notebook"
sns.set_context("notebook")

# Create bar plot
sns.catplot(x="Number of Siblings", y="Feels Lonely",
            data=survey_data, kind="bar")

# Show plot
plt.show()
```

```python
# Change the context to "talk"
sns.set_context("talk")

# Create bar plot
sns.catplot(x="Number of Siblings", y="Feels Lonely",
            data=survey_data, kind="bar")

# Show plot
plt.show()
```

```python
# Change the context to "poster"
sns.set_context("poster")

# Create bar plot
sns.catplot(x="Number of Siblings", y="Feels Lonely",
            data=survey_data, kind="bar")

# Show plot
plt.show()
```

**Note** : Setiap nama konteks memberikan saran Seaborn tentang kapan harus menggunakan skala plot yang diberikan (di kertas, di notebook iPython, dalam pembicaraan / presentasi, atau dalam sesi poster).

### Using a custom palette

Sejauh ini, kami telah melihat beberapa hal dalam dataset tanggapan survei dari kaum muda, termasuk penggunaan internet mereka, seberapa sering mereka mendengarkan orang tua mereka, dan berapa banyak dari mereka yang melaporkan merasa kesepian. Namun, satu hal yang belum kami lakukan adalah ringkasan dasar dari tipe orang yang menjawab survei ini, termasuk usia dan jenis kelamin mereka. Memberikan ringkasan dasar ini selalu merupakan praktik yang baik ketika berhadapan dengan dataset yang tidak dikenal.

Kode yang diberikan akan membuat box plot yang menunjukkan distribusi usia untuk responden pria dan wanita. Mari kita sesuaikan kode untuk menyesuaikan penampilan, kali ini menggunakan palet warna khusus.

```python
# Set the style to "darkgrid"
sns.set_style("darkgrid")

# Set a custom color palette
sns.set_palette(["#39A7D0", "#36ADA4"])

# Create the box plot of age distribution by gender
sns.catplot(x="Gender", y="Age", 
            data=survey_data, kind="box")

# Show plot
plt.show()
```

**Note** : Sepertinya usia rata-rata adalah sama untuk pria dan wanita, tetapi distribusi wanita cenderung lebih muda daripada pria.

## Adding titles and labels: Part 1

### FacetGrids vs. AxesSubplots

Dalam pelajaran terakhir, kami belajar bahwa fungsi plot Seaborn membuat dua jenis objek: objek `FacetGrid` dan objek `AxesSubplot`. Metode untuk menambahkan judul pada plot Anda akan berbeda tergantung pada jenis objeknya.

Dalam kode yang disediakan, kami telah menggunakan `relplot()` dengan dataset miles per galon untuk membuat scatter plot yang menunjukkan hubungan antara bobot mobil dan tenaga kuda. Scatter plot ini menggunakan nama variabel `g`. Mari kita mengidentifikasi jenis objek apa itu.

```python
# Load data
mpg = pd.read_csv("datasets/mpg.csv")
```

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Identify plot type
type_of_g = type(g)

# Print type
print(type_of_g)
```

**Question**

Kami baru saja melihat bahwa `sns.relplot()` membuat objek FacetGrid. Fungsi Seaborn lain apa yang membuat objek FacetGrid alih-alih objek AxesSubplot?

**Answer** : `catplot()` mendukung pembuatan subplot, sehingga ia menciptakan objek FacetGrid.

### Adding a title to a FacetGrid object

Sekarang kita tahu jenis objeknya, mari kita tambahkan judul pada plot ini.

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower")

# Show plot
plt.show()
```

**Note** : Sepertinya bobot mobil berkorelasi positif dengan tenaga kuda.

## Adding titles and labels: Part 2

### Adding a title and axis labels

Mari kita terus melihat mil per dataset galon. Kali ini kami akan membuat plot garis untuk menjawab pertanyaan: *Bagaimana rata-rata mil per galon yang dicapai oleh mobil berubah seiring waktu untuk masing-masing dari tiga tempat asal?* Untuk meningkatkan keterbacaan plot ini, kami akan menambahkan judul dan label sumbu yang lebih informatif.

Dalam kode yang disediakan, kami membuat plot garis menggunakan fungsi `lineplot()`. Perhatikan bahwa `lineplot()` tidak mendukung pembuatan subplot, jadi ia mengembalikan objek `AxesSubplot` alih-alih objek `FacetGrid`.

```python
# Load data
mpg_mean = pd.read_csv("datasets/mpg_mean.csv")
```

```python
# Grid off
sns.set_style("whitegrid", {'axes.grid' : False})

# Create line plot
g = sns.lineplot(x="model_year", y="mpg_mean", 
                 data=mpg_mean,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Show plot
plt.show()
```

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg_mean", 
                 data=mpg_mean,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Add x-axis and y-axis labels
g.set(xlabel="Car Model Year", ylabel="Average MPG")

# Show plot
plt.show()
```

**Note** : Mil rata-rata per galon yang dicapai meningkat dari waktu ke waktu untuk ketiga tempat asal, tetapi AS selalu lebih rendah daripada Eropa dan Jepang.

### Rotating x-tick labels

Kami membuat point plot yang menampilkan akselerasi rata-rata untuk mobil di masing-masing dari tiga tempat asal. Perhatikan bahwa variabel `"acceleration"` adalah waktu untuk berakselerasi dari 0 hingga 60 mil per jam, dalam detik. Nilai yang lebih tinggi menunjukkan akselerasi yang lebih lambat.

Mari gunakan plot ini untuk berlatih memutar label x-tick. Ingatlah bahwa fungsi untuk memutar label x-tick adalah fungsi Matplotlib mandiri dan bukan fungsi yang diterapkan pada objek plot itu sendiri.

```python
# Create point plot
sns.catplot(x="origin", 
            y="acceleration", 
            data=mpg, 
            kind="point", 
            join=False, 
            capsize=0.1)

# Rotate x-tick labels
plt.xticks(rotation=90)

# Show plot
plt.show()
```

**Note** : Karena nilai yang lebih tinggi menunjukkan akselerasi yang lebih lambat, sepertinya mobil dari Jepang dan Eropa memiliki akselerasi yang lebih lambat secara signifikan dibandingkan dengan AS.

## Putting it all together

### Box plot with subgroups

Dalam latihan ini, kita akan melihat dataset yang berisi tanggapan dari survei yang diberikan kepada kaum muda. Salah satu pertanyaan yang diajukan ke anak-anak muda adalah: *"Apakah Anda tertarik memiliki hewan peliharaan?"* Mari kita telusuri apakah distribusi usia mereka yang menjawab "ya" cenderung lebih tinggi atau lebih rendah daripada mereka yang menjawab "tidak", mengendalikan gender.

```python
# Set palette to "Blues"
sns.set_palette("Blues")

# Adjust to add subgroups based on "Interested in Pets"
g = sns.catplot(x="Gender",
                y="Age", data=survey_data, 
                kind="box", hue="Interested in Pets")

# Set title to "Age of Those Interested in Pets vs. Not"
g.fig.suptitle("Age of Those Interested in Pets vs. Not")

# Show plot
plt.show()
```

**Note** : Setelah mengendalikan gender, sepertinya distribusi usia orang-orang yang tertarik pada hewan peliharaan sama dengan mereka yang tidak.

### Bar plot with subgroups and subplots

Sekarang, kita akan menyelidiki apakah proporsi orang yang menyukai musik techno (`"Likes Techno"`) bervariasi berdasarkan gender (`"Gender"`) atau di mana mereka tinggal (`"Village - town"`). Latihan ini akan memberi kita kesempatan untuk mempraktikkan banyak hal yang telah kita pelajari selama kursus ini!

```python
# Set the figure style to "dark"
sns.set_style("dark")

# Adjust to add subplots per gender
g = sns.catplot(x="Village - town", y="Likes Techno", 
                data=survey_data, kind="bar",
                col="Gender")

# Add title and axis labels
g.fig.suptitle("Percentage of Young People Who Like Techno", y=1.02)
g.set(xlabel="Location of Residence", 
       ylabel="% Who Like Techno")

# Show plot
plt.show()
```
