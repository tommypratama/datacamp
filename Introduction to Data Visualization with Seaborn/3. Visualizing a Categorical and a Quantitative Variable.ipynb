# Visualizing a Categorical and a Quantitative Variable

Variabel kategorikal hadir di hampir setiap dataset, tetapi mereka sangat menonjol dalam data survei. Dalam bab ini, Anda akan belajar cara membuat dan menyesuaikan plot kategorikal seperti box plots, bar plots, count plots, dan point plots. Sepanjang jalan, Anda akan mengeksplorasi data survei dari anak muda tentang minat mereka, kebiasaan belajar siswa, dan pria dewasa tentang perasaan mereka tentang maskulinitas.

## Count plots and bar plots


### Count plots

Dalam latihan ini, kami akan kembali ke mengeksplorasi dataset kami yang berisi tanggapan atas survei yang dikirim kepada kaum muda. Kami mungkin curiga bahwa kaum muda menghabiskan banyak waktu di internet, tetapi berapa banyak yang mereka laporkan menggunakan internet setiap hari? Mari kita gunakan plot perhitungan untuk menjabarkan jumlah respons survei di setiap kategori dan kemudian mencari tahu apakah itu berubah berdasarkan usia.

Sebagai pengingat, untuk membuat plot hitungan, kami akan menggunakan fungsi `catplot()` dan menentukan nama variabel kategorikal yang akan dihitung (`x=____`), Pandas DataFrame yang akan digunakan (`data=____`), dan jenis plot (`kind="count"`).

```python
# Import library
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
# Set style
sns.set()
```

```python
# Load data
survey_data = pd.read_csv("datasets/young-people-survey.csv")
survey_data.sample(5)
```

```python
# Create count plot of internet usage
sns.catplot(x="Internet usage", data=survey_data, kind="count")

# Show plot
plt.show()
```

```python
# Change the orientation of the plot
sns.catplot(y="Internet usage", data=survey_data, kind="count")

# Show plot
plt.show()
```

```python
# Create column subplots based on age category
sns.catplot(y="Internet usage", data=survey_data,
            kind="count", col="Age Category")

# Show plot
plt.show()
```

**Note** : Sepertinya kebanyakan anak muda menggunakan internet selama beberapa jam setiap hari, berapapun usianya.

### Bar plots with percentages

Variabel `"Interested in Math"` adalah `True` jika orang yang dilaporkan tertarik atau sangat tertarik pada matematika, dan `False` sebaliknya. Berapa persentase kaum muda yang melaporkan minat dalam matematika, dan apakah ini berbeda berdasarkan gender? Mari kita gunakan plot bar untuk mencari tahu.


```python
# Create a bar plot of interest in math, separated by gender
sns.catplot(x="Gender", y="Interested in Math", data=survey_data, kind="bar")

# Show plot
plt.show()
```

**Note** : Ketika variabel-y adalah True/False, plot bar akan menampilkan persentase respons yang melaporkan True. Plot ini menunjukkan kepada kita bahwa pria melaporkan minat yang jauh lebih tinggi dalam matematika dibandingkan dengan wanita.

### Customizing bar plots

Dalam latihan ini, kami akan mengeksplorasi data dari siswa di sekolah menengah `student_data`. Variabel `"study_time"` mencatat setiap siswa melaporkan waktu belajar mingguan sebagai salah satu kategori berikut: `"<2 hours"`, `"2 to 5 hours"`, `"5 to 10 hours"`, atau `">10 hours"`. Apakah siswa yang melaporkan jumlah belajar yang lebih tinggi cenderung mendapatkan nilai akhir yang lebih baik? Mari kita bandingkan nilai akhir rata-rata di antara siswa di setiap kategori menggunakan plot bar.

```python
# Load data
student_data = pd.read_csv("datasets/student-alcohol-consumption.csv", index_col=0)
student_data.head()
```

```python
# Create bar plot of average final grade in each study category
sns.catplot(x="study_time", y="G3", data=student_data, kind="bar")

# Show plot
plt.show()
```

```python
# Rearrange the categories
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar", order=["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"])

# Show plot
plt.show()
```

```python
# Turn off the confidence intervals
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=["<2 hours", "2 to 5 hours", "5 to 10 hours", ">10 hours"], 
            ci=None)

# Show plot
plt.show()
```

**Note** : Siswa dalam sampel kami yang belajar lebih banyak memiliki nilai rata-rata yang sedikit lebih tinggi, tetapi itu bukan hubungan yang kuat.

## Box plots

### Create and interpret a box plot

Dalam latihan sebelumnya, kami mengeksplorasi hubungan antara belajar dan nilai akhir dengan menggunakan plot bar untuk membandingkan nilai akhir rata-rata (`"G3"`) di antara siswa dalam berbagai kategori `"study_time"`.

Dalam latihan ini, kami akan mencoba menggunakan tampilan box plot pada hubungan ini sebagai gantinya.

```python
# Specify the category ordering
study_time_order = ["<2 hours", "2 to 5 hours", 
                    "5 to 10 hours", ">10 hours"]

# Create a box plot and set the order of the categories
sns.catplot(x="study_time", y="G3", data=student_data, kind="box", order=study_time_order)

# Show plot
plt.show()
```

**Note** : Garis di tengah setiap kotak mewakili median.

### Omitting outliers

Sekarang mari kita gunakan dataset `student_data` untuk membandingkan distribusi nilai akhir (`"G3"`) antara siswa yang memiliki akses internet di rumah dan mereka yang tidak. Untuk melakukan ini, kita akan menggunakan variabel `"internet"`, yang merupakan indikator biner (yes/ o) apakah siswa memiliki akses internet di rumah.

Karena internet mungkin kurang dapat diakses di daerah pedesaan, kami akan menambahkan subkelompok berdasarkan tempat tinggal siswa. Untuk ini, kita dapat menggunakan variabel `"location"`, yang merupakan indikator apakah seorang siswa tinggal di lokasi perkotaan (`"Urban"`) atau pedesaan (`"Rural"`).

Sebagai pengingat, Anda dapat menghilangkan outlier dalam box plot dengan mengatur parameter `sym` sama dengan string kosong (`""`).

```python
# Create a box plot with subgroups and omit the outliers
sns.catplot(x="internet", y="G3", data=student_data, kind="box",
            sym="", hue="location")

# Show plot
plt.show()
```

**Note** : Nilai median sangat mirip antara masing-masing kelompok, tetapi penyebaran distribusi terlihat lebih besar di antara siswa yang memiliki akses internet.

### Adjusting the whiskers

Dalam pelajaran kami melihat bahwa ada beberapa cara untuk mendefinisikan whiskers dalam box plot. Dalam rangkaian latihan ini, kami akan terus menggunakan dataset `student_data` untuk membandingkan distribusi nilai akhir (`"G3"`) antara siswa yang berada dalam hubungan romantis dan yang tidak. Kami akan menggunakan variabel `"romantic"`, yang merupakan indikator yes/no dari apakah siswa berada dalam hubungan romantis.

Mari kita buat box plot untuk melihat hubungan ini dan mencoba berbagai cara untuk mendefinisikan whiskers.

```python
# Set the whiskers to 0.5 * IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box", whis=0.5)

# Show plot
plt.show()
```

```python
# Extend the whiskers to the 5th and 95th percentile
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[5, 95])

# Show plot
plt.show()
```

```python
# Set the whiskers at the min and max values
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0, 100])

# Show plot
plt.show()
```

**Note** : Nilai median adalah sama antara kedua kelompok ini, tetapi nilai maksimal lebih tinggi di antara siswa yang tidak memiliki hubungan romantis.

## Point plots


### Customizing point plots

Kali ini kita akan menggunakan point plot untuk menjawab pertanyaan: *Apakah kualitas hubungan keluarga siswa mempengaruhi jumlah ketidakhadiran siswa di sekolah?* Di sini, kita akan menggunakan variabel `"famrel"`, yang menggambarkan kualitas hubungan keluarga siswa dari 1 (sangat buruk) hingga 5 (sangat baik).

```python
# Create a point plot of family relationship vs. absences
sns.catplot(x="famrel", y="absences", data=student_data, kind="point")

# Show plot
plt.show()
```

```python
# Add caps to the confidence interval
sns.catplot(x="famrel", y="absences", data=student_data, kind="point", capsize=0.2)
        
# Show plot
plt.show()
```

```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences", data=student_data,
            kind="point", capsize=0.2, join=False)
            
# Show plot
plt.show()
```

**Note** : Sementara jumlah rata-rata absensi sedikit lebih kecil di antara siswa dengan hubungan keluarga berkualitas tinggi, interval kepercayaan yang besar memberi tahu kami bahwa kami tidak dapat memastikan ada hubungan yang sebenarnya di sini.

### Point plots with subgroups

Pertanyaan selanjutnya: apakah berada dalam hubungan romantis yang terkait dengan kehadiran di sekolah jauh lebih tinggi atau lebih rendah? Dan apakah asosiasi ini berbeda dengan sekolah yang dihadiri siswa? Mari cari tahu menggunakan point plot.

```python
# Create a point plot with subgroups
sns.catplot(x="romantic", y="absences", data=student_data, kind="point", hue="school")

# Show plot
plt.show()
```

```python
# Turn off the confidence intervals for this plot
sns.catplot(x="romantic", y="absences", data=student_data,
            kind="point", hue="school", ci=None)

# Show plot
plt.show()
```

```python
# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences", data=student_data,
            kind="point", hue="school", ci=None, estimator=np.median)

# Show plot
plt.show()
```

**Note** : Sepertinya siswa dalam hubungan romantis memiliki angka absensi rata-rata dan median yang lebih tinggi di sekolah dokter umum (GP), tetapi hubungan ini tidak berlaku untuk sekolah MS.
