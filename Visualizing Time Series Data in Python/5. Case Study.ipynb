Sebagai pengingat, Anda dapat mengekstrak tahun dari setiap tanggal dalam indeks menggunakan atribut `.index.year`.
````python
# Extract the month from the index of jobs
index_month = jobs.index.month

# Compute the mean unemployment rate for each month
jobs_by_month = jobs.groupby(index_month).mean()

# Plot the mean unemployment rate for each month
ax = jobs_by_month.plot(fontsize=10, linewidth=1)

# Set axis labels and legend
ax.set_xlabel('Month', fontsize=10)
ax.set_ylabel('Mean unemployment rate', fontsize=12)
ax.legend(bbox_to_anchor=(0.8, 0.6), fontsize=6)
plt.show()
````

````python
# Extract of the year in each date indices of the jobs DataFrame
index_year = jobs.index.year

# Compute the mean unemployment rate for each year
jobs_by_year = jobs.groupby(index_year).mean()

# Plot the mean unemployment rate for each year
ax = jobs_by_year.plot(fontsize=10, linewidth=1)

# Set axis labels and legend
ax.set_xlabel('Year', fontsize=10)
ax.set_ylabel('Mean unemployment rate', fontsize=10)
ax.legend(bbox_to_anchor=(0.1, 0.5), fontsize=10)
plt.show()
````

## Decompose time series data

### Apply time series decomposition to your dataset

Anda sekarang akan melakukan dekomposisi time series pada beberapa time series. Anda dapat mencapai ini dengan memanfaatkan Python dictionary untuk menyimpan hasil dari setiap penguraian time series.

Dalam latihan ini, Anda akan menginisialisasi dictionary kosong dengan satu set kurung kurawal, `{}`, gunakan perulangan `for` untuk beralih melalui kolom-kolom DataFrame dan terapkan dekomposisi time series untuk setiap time series. Setelah setiap dekomposisi time series, Anda menempatkan hasilnya di dictionary dengan menggunakan perintah `my_dict[key] = value`, di mana `my_dict` adalah dictionary Anda, `key` adalah nama dari kolom / time series, dan `value` adalah objek dekomposisi time series.

````python
import statsmodels.api as sm

# Initialize dictionary
jobs_decomp = {}

# Get the names of each time series in the DataFrame
jobs_names = jobs.columns

# Run time series decomposition on each time series of the DataFrame
for ts in jobs_names:
    ts_decomposition = sm.tsa.seasonal_decompose(jobs[ts])
    jobs_decomp[ts] = ts_decomposition
````

### Visualize the seasonality of multiple time series

Anda sekarang akan mengekstrak komponen `seasonality` dari `jobs_decomp` untuk memvisualisasikan musiman dalam time series ini. Perhatikan bahwa sebelum memplot, Anda harus mengonversi dictionary komponen `seasonality` menjadi DataFrame menggunakan fungsi `pd.DataFrame.from_dict()`.

````python
jobs_seasonal = {}

# Extract the seasonal values for the decomposition of each time series
for ts in jobs_names:
    jobs_seasonal[ts] = jobs_decomp[ts].seasonal
    
# Create a DataFrame from the jobs_seasonal dictionary
seasonality_df = pd.DataFrame.from_dict(jobs_seasonal)

# Remove the label for the index
seasonality_df.index.name = None

# Create a faceted plot of the seasonality_df DataFrame
seasonality_df.plot(subplots=True,
                   layout=(4, 4),
                   sharey=False,
                   fontsize=8,
                   linewidth=0.3,
                   legend=False,
                   figsize=(12, 5))

# Show plot
plt.show()
````

**Catatan** : Setiap time series dalam jobs DataFrame memiliki pola musiman yang sangat berbeda!
