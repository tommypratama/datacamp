# 2. Plotting 2D arrays

Bab ini menampilkan berbagai teknik untuk memvisualisasikan array dua dimensi. Ini termasuk penggunaan, presentasi, dan orientasi grid untuk mewakili fungsi dua variabel diikuti dengan diskusi plot pseudocolor, plot kontur, peta warna, histogram dua dimensi, dan gambar.

## Working with 2D arrays

### Generating meshes

Untuk memvisualisasikan data array dua dimensi, perlu dipahami bagaimana membuat dan memanipulasi array 2-D. Banyak plot Matplotlib mendukung array sebagai input dan khususnya, mereka mendukung array NumPy. Perpustakaan NumPy adalah sarana yang paling banyak didukung untuk mendukung array numerik di Python.

Dalam latihan ini, Anda akan menggunakan fungsi `meshgrid` di NumPy untuk menghasilkan array 2-D yang kemudian akan Anda visualisasikan menggunakan `plt.imshow()`. Cara paling sederhana untuk menghasilkan `meshgrid` adalah sebagai berikut:

<pre>
import numpy as np
Y,X = np.meshgrid(range(10),range(20))
</pre>

Ini akan membuat dua array dengan bentuk `(10,20)`, yang sesuai dengan 10 baris sepanjang sumbu Y dan 20 kolom di sepanjang sumbu X. Dalam latihan ini, Anda akan menggunakan `np.meshgrid()` untuk menghasilkan sampling 2-D reguler dari fungsi matematika.

```python
# Import numpy and matplotlib.pyplot
import numpy as np
import matplotlib.pyplot as plt

# Generate two 1-D arrays: u, v
u = np.linspace(-2, 2, 41)
v = np.linspace(-1, 1, 21)

# Generate 2-D arrays from u and v: X, Y
X,Y = np.meshgrid(u, v)

# Compute Z based on X and Y
Z = np.sin(3*np.sqrt(X**2 + Y**2)) 

# Display the resulting image with pcolor()
plt.pcolor(Z)
plt.show()

# Save the figure to 'sine_mesh.png'
plt.savefig('images/sine_mesh.png')
```

## Visualizing bivariate functions

### Contour & filled contour plots

Meskipun `plt.imshow()` atau `plt.pcolor()` sering digunakan untuk memvisualisasikan array 2-D secara keseluruhan, ada cara lain untuk memvisualisasikan data tersebut tanpa menampilkan semua nilai sampel yang tersedia. Salah satu opsi adalah menggunakan array untuk menghitung kontur yang divisualisasikan sebagai gantinya.

Dua jenis plot kontur yang didukung oleh Matplotlib adalah `plt.contour()` dan `plt.contourf()` di mana yang pertama menampilkan kontur sebagai garis dan yang terakhir ditampilkan mengisi area di antara kontur. Kedua perintah plot ini menerima array dua dimensi dari mana kontur yang sesuai dihitung.

Dalam latihan ini, Anda akan memvisualisasikan array 2-D berulang kali menggunakan `plt.contour()` dan `plt.contourf()`. Anda akan menggunakan `plt.subplot()` untuk menampilkan beberapa plot kontur pada gambar yang sama, menggunakan meshgrid `X`, `Y` sebagai sumbu. Sebagai contoh, `plt.contour(X, Y, Z)` menghasilkan peta kontur default dari array `Z`.

*Jangan lupa untuk memasukkan meshgrid di setiap plot untuk latihan ini!*

```python
# Generate a default contour map of the array Z
plt.subplot(2,2,1)
plt.contour(X, Y, Z)

# Generate a contour map with 20 contours
plt.subplot(2,2,2)
plt.contour(X, Y, Z, 20)

# Generate a default filled contour map of the array Z
plt.subplot(2,2,3)
plt.contourf(X, Y, Z)

# Generate a default filled contour map with 20 contours
plt.subplot(2,2,4)
plt.contourf(X, Y, Z, 20)

# Improve the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

**Note** : Seperti yang Anda lihat, dua plot kontur bagian bawah terisi, karena Anda menggunakan `plt.contourf()` alih-alih `plt.contour()`.

### Modifying colormaps

Saat menampilkan array 2-D dengan `plt.imshow()` atau `plt.pcolor()`, nilai-nilai array dipetakan ke warna yang sesuai. Himpunan warna yang digunakan ditentukan oleh colormap yang memetakan nilai ke warna dengan lancar, sehingga memudahkan untuk memahami struktur data secara sekilas.

Seringkali berguna untuk mengubah colormap dari colormap `'jet'` default yang digunakan oleh matplotlib. Colormap yang baik secara visual menyenangkan dan menyampaikan struktur data dengan tepat dan dengan cara yang masuk akal untuk aplikasi tersebut.

* Beberapa colormaps matplotlib memiliki nama unik seperti jet', 'coolwarm', 'magma' dan 'viridis'.
* Lainnya memiliki skema penamaan berdasarkan warna keseluruhan seperti 'Greens', 'Blues', 'Reds', dan 'Purples'
* Empat colormaps lain didasarkan pada musim, yaitu 'summer', 'autumn', 'winter' dan 'spring'

Dalam latihan ini, Anda akan menjelajahi empat colormaps yang berbeda menggunakan `plt.subplot()`. Anda akan menggunakan array pregenerated `Z` dan meshgrid `X`, `Y` untuk menghasilkan plot kontur yang diisi sama dengan empat peta warna yang berbeda. Pastikan juga menambahkan bilah warna ke setiap plot kontur yang diisi dengan `plt.colorbar()`.

```python
# Create a filled contour plot with a color map of 'viridis'
plt.subplot(2,2,1)
plt.contourf(X,Y,Z,20, cmap='viridis')
plt.colorbar()
plt.title('Viridis')

# Create a filled contour plot with a color map of 'gray'
plt.subplot(2,2,2)
plt.contourf(X,Y,Z,20, cmap='gray')
plt.colorbar()
plt.title('Gray')

# Create a filled contour plot with a color map of 'autumn'
plt.subplot(2,2,3)
plt.contourf(X,Y,Z,20, cmap='autumn')
plt.colorbar()
plt.title('Autumn')

# Create a filled contour plot with a color map of 'winter'
plt.subplot(2,2,4)
plt.contourf(X,Y,Z,20, cmap='winter')
plt.colorbar()
plt.title('Winter')

# Improve the spacing between subplots and display them
plt.tight_layout()
plt.show()
```

**Note** : Memilih peta warna yang baik dapat membuatnya secara signifikan lebih mudah untuk dengan cepat memahami struktur data Anda.

## Visualizing bivariate distributions

### Using hist2d()

Diberikan satu set pasangan terurut yang menggambarkan titik data, Anda dapat menghitung jumlah titik dengan nilai yang sama untuk membuat histogram dua dimensi. Ini mirip dengan histogram satu dimensi, tetapi ini menggambarkan variasi gabungan dari dua variabel acak dan bukan hanya satu.

Dalam matplotlib, satu fungsi untuk memvisualisasikan histogram 2-D adalah `plt.hist2d()`

* Anda menentukan koordinat titik menggunakan `plt.hist2d(x, y)` dengan asumsi `x` dan `y` adalah dua vektor dengan panjang yang sama.
* Anda dapat menentukan jumlah bins dengan argumen `bins=(nx, ny)` di mana `nx` adalah jumlah bins untuk digunakan dalam arah horizontal dan `ny` adalah jumlah bins untuk digunakan dalam arah vertikal.
* Anda dapat menentukan wilayah persegi panjang di mana sampel dihitung dalam membangun histogram 2D. Parameter opsional yang diperlukan adalah `range=((xmin, xmax), (ymin, ymax))`
* `xmin` dan `xmax` adalah masing-masing batas bawah dan atas untuk variabel pada sumbu x.
* `ymin` dan `ymax` adalah masing-masing batas bawah dan atas untuk variabel pada sumbu y. Perhatikan bahwa argumen opsional `range` dapat menggunakan *nested tuples* atau *lists*.

Dalam latihan ini, Anda akan menggunakan beberapa data dari kumpulan data `auto-mpg`. Ada dua array `mpg` dan `hp` yang masing-masing berisi mil per galon dan peringkat daya kuda.

```python
# Import pandas
import pandas as pd

# Assign url
file_autompg = 'https://assets.datacamp.com/production/repositories/558/datasets/1a03987ad77b38d61fc4c692bf64454ddf345fbe/auto-mpg.csv'
# Load data
autompg = pd.read_csv(file_autompg)
autompg.head()
```

```python
# Create array
hp = np.array(autompg['hp'])
mpg = np.array(autompg['mpg'])
```

```python
# Generate a 2-D histogram
plt.hist2d(hp, mpg, bins=(20, 20), range=((40,235), (8, 48)))

# Add a color bar to the histogram
plt.colorbar()

# Add labels, title, and display the plot
plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hist2d() plot')
plt.show()
```

**Note** : Seperti yang Anda harapkan, mobil dengan `hp` lebih tinggi memiliki `mpg` lebih rendah.

### Using hexbin()

Fungsi `plt.hist2d()` menggunakan bins persegi panjang untuk membuat histogram dua dimensi. Sebagai alternatif, fungsi `plt.hexbin()` menggunakan bins *hexagonal*. Algoritma yang mendasari (berdasarkan [artikel ini dari 1987](http://amstat.tandfonline.com/doi/abs/10.1080/01621459.1987.10478445)) membangun tesselation heksagonal dari wilayah planar dan mengumpulkan poin di dalam bins hexagonal.

Argumen opsional `gridsize` (default 100) memberikan jumlah segi enam (*hexagons*) melintasi arah-x yang digunakan dalam ubin heksagonal. Jika ditentukan sebagai lists atau tuple dengan panjang dua, ukuran `gridsize` memperbaiki jumlah segi enam dalam arah x dan y masing-masing dalam ubin.

Parameter opsional `extent=(xmin, xmax, ymin, ymax)` menentukan wilayah persegi panjang yang dicakup oleh ubin heksagonal. Dalam hal itu, `xmin` dan `xmax` adalah masing-masing batas bawah dan atas untuk variabel pada sumbu x, `ymin` dan `ymax` masing-masing adalah batas bawah dan atas untuk variabel pada sumbu y.

Kali ini, Anda akan menggunakan `plt.hexbin()` untuk memvisualisasikan histogram dua dimensi.

```python
# Generate a 2d histogram with hexagonal bins
plt.hexbin(hp, mpg, gridsize=(15, 12), extent=(40,235,8,48))

# Add a color bar to the histogram
plt.colorbar()

# Add labels, title, and display the plot
plt.xlabel('Horse power [hp]')
plt.ylabel('Miles per gallon [mpg]')
plt.title('hexbin() plot')
plt.show()
```

**Note** : Terkadang, Anda mungkin lebih suka menggunakan bins heksagonal, seperti dalam plot ini, daripada bins persegi panjang.

## Working with images

### Loading, examining images

Gambar berwarna seperti foto, mengandung intensitas saluran warna merah, hijau dan biru.

* Saluran warna dapat diekstraksi dengan Numpy array slicing.

### Pseudocolor plot from image data

Data gambar tersedia dalam berbagai bentuk dan tidak selalu sesuai untuk menampilkan saluran yang tersedia dalam ruang RGB. Dalam banyak situasi, suatu gambar dapat diproses dan dianalisis dalam beberapa cara sebelum divisualisasikan dalam pseudocolor, juga dikenal sebagai *'false' color*.

Dalam latihan ini, Anda akan melakukan analisis sederhana menggunakan gambar yang menunjukkan seorang astronot dilihat dari luar angkasa. Alih-alih hanya menampilkan gambar, Anda akan menghitung intensitas total di saluran merah, hijau dan biru. Hasilnya adalah array dua dimensi tunggal yang akan Anda tampilkan menggunakan `plt.imshow()` dengan colormap `'gray'`.

```python
# Load the image into an array: img
img = plt.imread('images/480px-Astronaut-EVA.jpg')

# Print the shape of the image
print(img.shape)

# Compute the sum of the red, green and blue channels: intensity
intensity = img.sum(axis=2)

# Print the shape of the intensity
print(intensity.shape)

# Display the intensity with a colormap of 'gray'
plt.imshow(intensity, cmap='gray')

# Add a colorbar
plt.colorbar()

# Hide the axes and show the figure
plt.axis('off')
plt.show()
```

**Note** : Perhatikan bagaimana array sekarang 2D.

### Extent and aspect

Saat menggunakan `plt.imshow()` untuk menampilkan array, perilaku default adalah menjaga pixel square sehingga rasio tinggi terhadap lebar dari output sesuai dengan rasio yang ditentukan oleh bentuk array. Selain itu, secara default, sumbu x dan y ditandai dengan jumlah sampel di setiap arah.

Rasio lebar ditampilkan dengan tinggi dikenal sebagai aspek gambar dan rentang yang digunakan untuk label sumbu x dan y dikenal sebagai `extent` gambar. Nilai default `aspect` adalah `'auto'` membuat piksel persegi dan luasan dihitung secara otomatis dari bentuk array jika tidak ditentukan sebaliknya.

Dalam latihan ini, Anda akan menyelidiki cara mengatur opsi-opsi ini secara eksplisit dengan memplot gambar yang sama dalam grid subplot 2 x 2 dengan perbedaan `aspect` dan opsi `extent`.

```python
# Load the image into an array: img
img = plt.imread('images/480px-Astronaut-EVA.jpg')

# Specify the extent and aspect ratio of the top left subplot
plt.subplot(2,2,1)
plt.title('extent=(-1,1,-1,1),\naspect=0.5') 
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=0.5)

# Specify the extent and aspect ratio of the top right subplot
plt.subplot(2,2,2)
plt.title('extent=(-1,1,-1,1),\naspect=1')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=1)

# Specify the extent and aspect ratio of the bottom left subplot
plt.subplot(2,2,3)
plt.title('extent=(-1,1,-1,1),\naspect=2')
plt.xticks([-1,0,1])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-1,1,-1,1), aspect=2)

# Specify the extent and aspect ratio of the bottom right subplot
plt.subplot(2,2,4)
plt.title('extent=(-2,2,-1,1),\naspect=2')
plt.xticks([-2,-1,0,1,2])
plt.yticks([-1,0,1])
plt.imshow(img, extent=(-2,2,-1,1), aspect=2)

# Improve spacing and display the figure
plt.tight_layout()
plt.show()
```

### Rescaling pixel intensities

Terkadang, gambar dengan kontras rendah dapat ditingkatkan dengan mengubah intensitasnya. Misalnya, gambar Hawkes Bay, Selandia Baru ini tidak memiliki nilai piksel dekat 0 atau dekat 255 (batas intensitas yang valid). (aslinya oleh Phillip Capper, dimodifikasi oleh Pengguna: Konstable, via Wikimedia Commons, [CC BY 2.0](http://creativecommons.org/licenses/by/2.0))

![](https://assets.datacamp.com/production/course_1761/datasets/640px-Unequalized_Hawkes_Bay_NZ.jpg)

Untuk latihan ini, Anda akan melakukan *rescaling* sederhana (ingat, gambar adalah array NumPy) untuk menerjemahkan dan meregangkan intensitas piksel sehingga intensitas gambar baru mengisi rentang dari 0 hingga 255.

```python
# Load the image into an array: image
image = plt.imread('images/640px-Unequalized_Hawkes_Bay_NZ.jpg')

# Extract minimum and maximum values from the image: pmin, pmax
pmin, pmax = image.min(), image.max()
print("The smallest & largest pixel intensities are %d & %d." % (pmin, pmax))

# Rescale the pixels: rescaled_image
rescaled_image = 256*(image - pmin) / (pmax - pmin)
print("The rescaled smallest & largest pixel intensities are %.1f & %.1f." % 
      (rescaled_image.min(), rescaled_image.max()))

# Display the rescaled image
plt.title('rescaled image')
plt.axis('off')
plt.imshow(rescaled_image)

plt.show()
```

**Note** : Seperti yang Anda lihat, gambar yang dire-skal memiliki intensitas piksel antara 0 dan 256.
